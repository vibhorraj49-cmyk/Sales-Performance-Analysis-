```python
import pandas as pd

df = pd.read_csv("superstore_final_dataset (1).csv", encoding='latin1')

print(df.head())
print("Shape:", df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Order_Date'] = pd.to_datetime(
      df['Order_Date'],
      dayfirst=True,
      errors='coerce'
)

df['Ship_Date'] = pd.to_datetime(
      df['Ship_Date'],
      dayfirst=True,
      errors='coerce'
)
```

```python
df = df.dropna()
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
total_revenue = df['Sales'].sum()

print("Total Revenue:", total_revenue)
```

```python
total_orders = df['Order_ID'].nunique()

print("Total Orders:", total_orders)
```

```python
average_order_value = (
      df['Sales'].sum()
          /
              df['Order_ID'].nunique()
              )

print("Average Order Value:", average_order_value)
```

```python
category_sales = df.groupby(
      'Category'
      )['Sales'].sum()

print(category_sales)
```

```python
region_sales = df.groupby(
      'Region'
      )['Sales'].sum()

print(region_sales)
```

```python
top_products = df.groupby(
      'Product_Name'
      )['Sales'].sum().sort_values(
          ascending=False
          )

print(top_products.head(10))
```

```python
monthly_sales = df.groupby(
      df['Order_Date'].dt.to_period('M')
      )['Sales'].sum()
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(
    figsize=(12,5)
    )

plt.title("Monthly Sales Trend")

plt.ylabel("Sales")

plt.show()
```

```python
region_sales.plot(
      kind='bar',
          figsize=(8,5)
          )

plt.title("Sales by Region")

plt.ylabel("Revenue")

plt.show()
```

```python
category_sales = df.groupby('Category')[
      'Sales'
      ].sum().sort_values(ascending=False)

print(category_sales)
```

```python
category_sales.plot(
      kind='bar',
          figsize=(8,5)
          )

plt.title("Sales by Category")

plt.ylabel("Revenue")

plt.show()
```

```python
top_products = df.groupby(
      'Sub_Category'
      )['Sales'].sum().sort_values(
          ascending=False
          )

print(top_products.head(10))
```

```python
top_products.head(10).plot(
      kind='bar',
          figsize=(10,5)
          )

plt.title("Top Performing Products")

plt.ylabel("Sales")

plt.show()
```

```python
worst_products = df.groupby(
      'Sub_Category'
      )['Sales'].sum().sort_values(
          ascending=True
          )

print(worst_products.head(10))
```

```python
import matplotlib.pyplot as plt

worst_products.head(10).plot(
    kind='bar',
        figsize=(10,5)
        )

plt.title("Lowest Performing Sub Categories")

plt.ylabel("Sales")

plt.show()
```

```python
# Ensure your order date is in datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)

# Extract month name
df['Month'] = df['Order_Date'].dt.month_name()
```

```python
seasonality = df.groupby('Month')['Sales'].sum()

print(seasonality)
```

```python
import matplotlib.pyplot as plt

seasonality.plot(
    kind='bar',
        figsize=(10,5)
        )

plt.title("Seasonality Analysis - Monthly Sales")
plt.ylabel("Sales")
plt.show()
```

```python
df['Shipping Days'] = (
      df['Ship_Date']
          - df['Order_Date']
          ).dt.days

avg_shipping = df['Shipping Days'].mean()

print(avg_shipping)
```

```python
segment_sales = df.groupby('Segment')[
      'Sales'
      ].sum()

print(segment_sales)
```

```python
segment_sales.plot(
      kind='pie',
          autopct='%1.1f%%',
              figsize=(8,8)
              )

plt.title("Sales by Customer Segment")

plt.ylabel("")

plt.show()
```

```python
pivot_table = df.pivot_table(
      values='Sales',
          index='Region',
              columns='Category',
                  aggfunc='sum'
                  )

print(pivot_table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.heatmap(
    pivot_table,
        annot=True,
            cmap='coolwarm'
            )

plt.title("Regional Sales Heatmap")

plt.show()
```

```python
correlation = df[
      ['Sales', 'Postal_Code', 'Shipping Days']
      ].corr()

print(correlation)
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
        annot=True,
            cmap='Blues'
            )

plt.title("Correlation Matrix")

plt.show()
```
